```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
print(x)
```

```python
plt.plot(x,y) # creating a empty figure
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sin values")
plt.title("sin wave")
plt.show()
```

```python
x=np.linspace(-10,10, 20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'b*')
plt.show()
```

```python
x=np.linspace( -5 ,5 ,50)
plt.plot(x, np.cos(x), 'g-')
plt.plot(x,np.sin(x),'r--')
plt.show()
```

```python
fig=plt.figure() # creating a empty figure
ax=fig.add_axes([0,0,1,1])# adding axes to them
languages=['English','French','Spanish','Latin','Newari']
peoples=[100,230,150,135,345]
ax.bar(languages,peoples)
plt.xlabel("languages")
plt.ylabel("peoples")
plt.title("Languages")
plt.show()
```

```python
fig= plt.figure()
ax= fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','Newari']
peoples=[100,230,150,135,345]
ax.pie(peoples,labels=languages,autopct='%1.1f%%')
plt.show()
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax2=fig2.add_axes([0,0,1,1])
ax2.scatter(x,y,color='r')
ax2.scatter(x,z,color='g')
plt.show()

```

3D scatter plot


```python
fig4=plt.figure()
ax4=plt.axes(projection='3d')
z=20* np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax4.scatter(x,y,z,c=z,cmap='Greens')
plt.show()
```
