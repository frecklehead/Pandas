```python
import seaborn as sns
import matplotlib.pyplot as plt
#importing some builtin datasets in seaborn named tips.
tips=sns.load_dataset('tips')

```

```python
tips.head()
```

```python
tips['time']
```

```python
#visualizing the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='sex',style='sex',style_order=None,size='size',sizes=(20,200))
```

```python
#setting theme for sns
sns.set_theme(style='darkgrid')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='sex',hue='time',style='smoker',size='size')
```

```python
sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.show()
```

```python
sns.histplot(tips["total_bill"],bins=20)
plt.show()
```

```python
sns.countplot(data=tips, x="day", hue="smoker")
plt.show()
```

```python
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex")
plt.show()
```
